### Model evaluator

```python
from sklearn.metrics import f1_score, recall_score, confusion_matrix

def evaluate_classification(y_true, y_pred):
    """
    Function to evaluate a classification model.

    Args:
    - y_true (array-like): True labels.
    - y_pred (array-like): Predicted labels.

    Returns:
    - f1 (float): F1 score.
    - recall (float): Recall score.
    - cm (array): Confusion matrix.
    """
    # Calculate F1 score
    f1 = f1_score(y_true, y_pred)
    
    # Calculate recall
    recall = recall_score(y_true, y_pred)
    
    # Calculate confusion matrix
    cm = confusion_matrix(y_true, y_pred)
    
    return f1, recall, cm

# Example usage:
# Assuming y_test and y_pred are available from a classification model
# Evaluate the model
# f1_score, recall_score, confusion_matrix = evaluate_classification(y_test, y_pred)

# # Output the evaluation metrics
# print("F1 Score:", f1_score)
# print("Recall Score:", recall_score)
# print("Confusion Matrix:\n", confusion_matrix)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

model_log = LogisticRegression()

# Train the model on the training data
model_log.fit(X_train, y_train)

# Predict on the testing data
y_pred_log = model_log.predict(X_test)

# Calculate accuracy
accuracy_log = accuracy_score(y_test, y_pred_log)
print("Accuracy:", accuracy_log)
```

```python
f1,recall,conf=evaluate_classification(y_test,y_pred_log)
```
